```python
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook as tqdm
import matplotlib.pyplot as plt
import re
import os
%matplotlib inline
```

```python
tickers = set()
for file in os.listdir('./s&p/'):
    try:
        limits = [(m.start(),m.end()) for m in re.finditer('\([^\)]+\)', file)][0]
        tickers.add(file[limits[0]+1:limits[1]-1])
    except:
        pass
```

```python
len(tickers)
```

```python
balance_sheets = []

for ticker in tqdm(tickers):
    try:
        local = pd.read_csv('./s&p/Balance Sheet({}).csv'.format(ticker), sep=';')
        local.rename(columns={'Breakdown': ticker}, inplace=True)
        local = local[local.columns[:2]].set_index(ticker)
        local = dict(local[local.columns[0]])
        local['Ticker'] = ticker
        balance_sheets.append(local)
    except FileNotFoundError:
        pass
```

```python
cashflow = []

for ticker in tqdm(tickers):
    try:
        local = pd.read_csv('./s&p/Cash Flow({}).csv'.format(ticker), sep=';')
        local.rename(columns={'Breakdown': ticker}, inplace=True)
        local = local[local.columns[:2]].set_index(ticker)
        local = dict(local[local.columns[0]])
        local['Ticker'] = ticker
        cashflow.append(local)
    except FileNotFoundError:
        pass
```

```python
income = []

for ticker in tqdm(tickers):
    try:
        local = pd.read_csv('./s&p/Income Statement({}).csv'.format(ticker), sep=';')
        local.rename(columns={'Breakdown': ticker}, inplace=True)
        local = local[local.columns[:2]].set_index(ticker)
        local = dict(local[local.columns[0]])
        local['Ticker'] = ticker
        income.append(local)
    except FileNotFoundError:
        pass
```

```python
stats = []

for ticker in tqdm(tickers):
    try:
        local = pd.read_csv('./s&p/Statistics({}).csv'.format(ticker), sep=';', header=None)
        local.rename(columns={0:ticker}, inplace=True)
        local = local.set_index(ticker)
        local.loc['Ticker'] = ticker
        stats.append(dict(local[1]))
    except FileNotFoundError:
        pass
```

```python
income = pd.DataFrame(income)
balance = pd.DataFrame(balance_sheets)
cashflow = pd.DataFrame(cashflow)
stats = pd.DataFrame(stats)
```

```python
income.set_index('Ticker', inplace=True)
balance.set_index('Ticker', inplace=True)
cashflow.set_index('Ticker', inplace=True)
stats.set_index('Ticker', inplace=True)
```

```python
stats_remove = ['Fiscal Year Ends','Last Split Date','Last Split Factor',
                'Ex-Dividend Date', 'Dividend Date','Forward Annual Dividend Rate', 
                'Forward Annual Dividend Yield','Shares Short ','Short Ratio ',
                'Shares Short (Jan 14, 2020)','Short Ratio (Jan 14, 2020)','Short % of Float (Jan 14, 2020)',
                'Short % of Shares Outstanding (Dec 30, 2019)','Shares Short (prior month Nov 28, 2019)',
                'Short % of Float ','Short % of Shares Outstanding ','Shares Short (prior month )',
                'Short % of Shares Outstanding (Jan 14, 2020)','Shares Short (prior month Dec 12, 2019)',
                'Shares Short (Dec 30, 2019)','Short Ratio (Dec 30, 2019)','Short % of Float (Dec 30, 2019)',
                'Most Recent Quarter','52 Week High','52 Week Low','50-Day Moving Average',
                '200-Day Moving Average','Shares Short (Jan 30, 2020)','Short Ratio (Jan 30, 2020)',
                'Short % of Float (Jan 30, 2020)','Short % of Shares Outstanding (Jan 30, 2020)',
                'Shares Short (prior month Dec 30, 2019)','Net Income','Revenue','Gross Profit',
                'Operating Cash Flow'
                ]
for col in stats_remove:
    try:
        stats.drop(col, axis=1, inplace=True)
    except KeyError:
        pass
```

```python
def to_num(s):
    if type(s)==str:
        if s[-1]=='T':
            return float(s[:-1]) * 1e12
        elif s[-1]=='B':
            return float(s[:-1]) * 1e9
        elif s[-1]=='M':
            return float(s[:-1]) * 1e9
        else:
            return np.nan
    else:
        return np.nan

def get_percent(s):
    if type(s)==str:
        try: return float(s[:-1])/100
        except: return np.nan
    else:
        return np.nan

def get_float(s):
    if type(s)==str:
        try: return float(s.replace(',',''))
        except: return np.nan
    else:
        return np.nan
```

```python
stats['Market Cap (intraday)'] = stats['Market Cap (intraday)'].apply(to_num)
stats['Enterprise Value'] = stats['Enterprise Value'].apply(to_num)
stats['Forward P/E'] = stats['Forward P/E'].apply(float)
stats['PEG Ratio (5 yr expected)'] = stats['PEG Ratio (5 yr expected)'].apply(float)
stats['Price/Sales'] = stats['Price/Sales'].apply(float)
stats['Price/Book'] = stats['Price/Book'].apply(float)
stats['Enterprise Value/Revenue'] = stats['Enterprise Value/Revenue'].apply(get_float)
stats['Enterprise Value/EBITDA'] = stats['Enterprise Value/EBITDA'].apply(get_float)
stats['Beta (5Y Monthly)'] = stats['Beta (5Y Monthly)'].apply(get_float)
stats['52-Week Change'] = stats['52-Week Change'].apply(get_percent)
stats['S&P500 52-Week Change'] = stats['S&P500 52-Week Change'].apply(get_percent)
stats['% Held by Insiders'] = stats['% Held by Insiders'].apply(get_percent)
stats['% Held by Institutions'] = stats['% Held by Institutions'].apply(get_percent)
stats['Trailing Annual Dividend Yield'] = stats['Trailing Annual Dividend Yield'].apply(get_percent)
stats['Payout Ratio'] = stats['Payout Ratio'].apply(get_percent)
stats['Trailing Annual Dividend Rate'] = stats['Trailing Annual Dividend Rate'].apply(get_float)
stats['5 Year Average Dividend Yield'] = stats['5 Year Average Dividend Yield'].apply(get_float)
stats['Profit Margin'] = stats['Profit Margin'].apply(get_percent)
stats['Operating Margin'] = stats['Operating Margin'].apply(get_percent)
stats['Return on Assets'] = stats['Return on Assets'].apply(get_percent)
stats['Return on Equity'] = stats['Return on Equity'].apply(get_percent)
stats['Revenue Per Share'] = stats['Revenue Per Share'].apply(get_float)
stats['Quarterly Revenue Growth'] = stats['Quarterly Revenue Growth'].apply(get_percent)
stats['EBITDA'] = stats['EBITDA'].apply(to_num)
stats['Net Income Avi to Common'] = stats['Net Income Avi to Common'].apply(to_num)
stats['Diluted EPS'] = stats['Diluted EPS'].apply(get_float)
stats['Quarterly Earnings Growth'] = stats['Quarterly Earnings Growth'].apply(get_percent)
stats['Total Cash'] = stats['Total Cash'].apply(to_num)
stats['Total Cash Per Share'] = stats['Total Cash Per Share'].apply(get_float)
stats['Total Debt'] = stats['Total Debt'].apply(to_num)
stats['Total Debt/Equity'] = stats['Total Debt/Equity'].apply(get_float)
stats['Current Ratio'] = stats['Current Ratio'].apply(get_float)
stats['Book Value Per Share'] = stats['Book Value Per Share'].apply(get_float)
stats['Levered Free Cash Flow'] = stats['Levered Free Cash Flow'].apply(to_num)
```

```python
stats['Trailing P/E'] = stats['Trailing P/E'].apply(get_float)
```
