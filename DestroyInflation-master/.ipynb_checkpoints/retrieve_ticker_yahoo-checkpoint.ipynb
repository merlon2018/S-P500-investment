```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common import exceptions
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd
import os

options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
driver = webdriver.Chrome("../chromedriver")

base_url = "https://finance.yahoo.com/quote/{}/history?p={}"

tickers = pd.read_csv("info_df.csv", index_col=0, sep='\t')
tickers = list(tickers.Ticker)

for ticker in tickers:
    # Get the URL of the ticker
    url = base_url.format(ticker, ticker)
    driver.get(url)
    sleep(3.)

    # Check of Accept button
    buttons = driver.find_elements_by_xpath("//button[@type='submit']")
    if len(buttons)>0:
        for button in buttons:
            s = button.get_attribute('innerHTML')
            print(s)
            if 'OK' in s or "J'accepte" in s:
                button.click()
                sleep(5.)
                break

    # Setting the max date to retrieve from
    div = driver.find_element_by_xpath("//section/div/div/div/div/div")
    div.click()
    sleep(1.)

    max_btn = driver.find_element_by_xpath("//button[@data-value='MAX']")
    max_btn.click()
    sleep(1.)

    buttons = driver.find_elements_by_xpath("//button")
    for button in buttons:
        if 'Apply' in str(button.get_attribute('innerHTML')):
            button.click()
            sleep(3.)
            break

    # Downloading from direct request link
    link = ""
    a_list = driver.find_elements_by_xpath("//a")
    for a in a_list:
        if 'Download Data' in str(a.get_attribute('innerHTML')):
            html = BeautifulSoup(str(a.get_attribute('outerHTML')), 'html.parser')
            link = html.find('a')['href']
            break
    driver.get(link)
    sleep(3.)

    # Putting the data in the right folder
    os.system('cp ../../Tlchargements/{}.csv ./stock_data/'.format(ticker))
    os.system('rm ../../Tlchargements/{}.csv'.format(ticker))
```
