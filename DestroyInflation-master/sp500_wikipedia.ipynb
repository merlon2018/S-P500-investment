```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common import exceptions
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd

driver = webdriver.Chrome("/Users/macbookpro_anas/Desktop/Machine-learning/drivers/chromedriver")
driver.get("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")
sleep(3.)

driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')

html = BeautifulSoup(driver.page_source)
tbody = html.find("tbody")
rows = tbody.find_all('tr')

tickers = []
names = []
sectors = []
subsectors = []
ciks = []
years = []

for row in rows:
    tds = row.find_all('td')
    tickers.append(str(tds[0].contents[0].contents[0]))
    names.append(str(tds[1].contents[0].contents[0]))
    sectors.append(str(tds[3].contents[0]))
    subsectors.append(str(tds[4].contents[0]))

info_df = pd.DataFrame([tickers, names, sectors, subsectors, ciks, years],
                        index=['Ticker','Name','Sector','Subsector','CIK','Year']).T

info_df.to_csv("info_df.csv", sep="\t")
```
